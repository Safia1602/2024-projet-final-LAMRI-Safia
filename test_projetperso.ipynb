# Quand les dialectes et langues officielles s’entrelacent : Une exploration des choix linguistiques

# Intoduction 

Les langues et les dialectes occupent une place centrale dans l'identité culturelle et sociale des individus, agissant non seulement comme des marqueurs d'appartenance mais aussi comme des instruments de communication influençant la manière dont les individus sont perçus dans la société. Dans un monde marqué par une diversité linguistique croissante et la coexistence de multiples langues, notamment à travers les migrations internes et internationales, les choix linguistiques, tels que le code-switching, révèlent des dynamiques complexes. Ces choix sont façonnés par une multiplicité de facteurs sociaux, géographiques et personnels. Parmi eux, l'utilisation des dialectes, souvent perçus comme des symboles d’identité culturelle, soulève des questions sur leur statut et leur perception dans divers contextes sociaux, notamment dans les environnements formels comme les activités administratives ou professionnelles, où certains dialectes peuvent être stigmatisés.

Le présent projet explore la manière dont les individus qui parlent des dialectes, tels que le dialecte algérien, marocain ou tunisien, ajustent leurs comportements linguistiques selon leur contexte de vie. À partir de données collectées auprès de jeunes adultes provenant de divers pays (principalement de l'Algérie et de la France), ce travail vise à analyser les relations entre l'usage des dialectes et les perceptions sociales, notamment les préjugés et stéréotypes relatifs à la langue. En croisant ces informations avec des données concernant la nationalité et le pays de résidence des participants, cette étude cherche à comprendre si ces facteurs influencent la fréquence et les contextes d'utilisation des dialectes, et dans quelle mesure ils sont liés à des perceptions de discrimination ou de rejet.


# Problématique: 

Dans un contexte de mondialisation croissante et d'urbanisation, les dynamiques linguistiques deviennent de plus en plus complexes. Le code-switching est un phénomène fréquent dans les interactions sociales, et sa fréquence et ses modalités varient selon les contextes. À travers l'examen de la relation entre les dialectes et les langues officielles, cette étude cherche à comprendre comment le code-switching et l'usage des dialectes sont influencés par des facteurs tels que la situation sociale (formelle ou informelle), la nationalité, et le pays de résidence. Cette recherche se pose la question suivante : 

Dans quelle mesure les choix linguistiques des participants, notamment l'utilisation des dialectes et le recours au code-switching, sont-ils modulés par ces facteurs contextuels et sociétaux ?

### Prédictions: 
- Les participants utilisent le code-switching en fonction des situations sociales.
- Les participants qui parlent un dialecte ajustent leur usage de celui-ci selon les situations, qu'elles soient formelles (activités administratives) ou informelles (à la maison, rêver).
- L'utilisation des dialectes varie selon la nationalité des participants.
- L'utilisation des dialectes diffère également selon le pays de résidence des participants.
- Les préjugés et stéréotypes liés aux dialectes dépendent du pays de résidence des participants.

# 3. Data manipulation

```python
# Supprimer les 17 premières colonnes
df = df.iloc[:, 17:]
# Réinitialiser les index
df = df.reset_index(drop=True)
# Afficher les premières lignes du DataFrame
df.head()
```

```python
# Identifier les colonnes de référence (Q8) et de contexte (Q12)
q8_columns = [col for col in df.columns if "Q8" in col]  # Colonnes avec les langues
q12_columns = [col for col in df.columns if "Q12" in col]  # Colonnes des situations

#  Créer une fonction pour remplacer les références par les langues
def remplacer_langue(valeur, row):
    for i, q8_col in enumerate(q8_columns):
        ref = f"Langue {i+1}" 
        if valeur == ref:  # Si la valeur dans Q12 correspond à une référence
            return row[q8_col]  # Retourner la langue correspondante
    return valeur  # Si aucune correspondance, conserver la valeur d'origine

# Appliquer la fonction sur les colonnes Q12
for col in q12_columns:
    df[col] = df.apply(lambda row: remplacer_langue(row[col], row), axis=1)

df[q12_columns].head()
```

```python
# Liste des colonnes Q12 représentant les différentes situations
questions = ['Q12_1', 'Q12_2', 'Q12_3', 'Q12_4', 'Q12_5', 'Q12_6', 'Q12_7', 'Q12_8']

# Liste des dialectes à garder
dialectes_valides = ['Dialecte algérien', 'Dialecte marocain', 'Dialecte tunisien', 'Catalan', 'Tagalog']

# Filtrer les lignes où au moins une des colonnes Q12 contient un dialecte valide
df_dialecte_max = df[df[questions].apply(lambda row: row.isin(dialectes_valides).any(), axis=1)]

print(df_dialecte_max[questions].head())
```

# 5. Test statistiques 

#### 1. Utilisation des dialectes dans des situations informelles, telles que dans le cadre familial

```python
from scipy.stats import chi2_contingency

# Create a contingency table of frequencies between two variables 
contingency_table = pd.crosstab(df['Q12_1_cat'], df['Q7'])

# Perform the chi-squared test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Display the results
print("Chi-squared Test:")
print("Chi-squared Statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of Freedom:", dof)
print("Table of Expected Frequencies:\n", expected)
```

### Interprétation: 
Avec un p-value inférieur à 0.05, le test du chi-carré suggère qu'il existe une association statistiquement significative entre l'utilisation d'un dialecte dans une situation informelle à la maison et les variables étudiées. 

#### 2. Utilisation des dialectes dans des situations formelles, telles que dans les activités administratifs

```python

# Create a contingency table of frequencies between the variables 'Q12_4_cat' and 'Q7'
contingency_table = pd.crosstab(df['Q12_4_cat'], df['Q7'])

# Perform the chi-squared test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Display the results
print("Chi-squared Test:")
print("Chi-squared Statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of Freedom:", dof)
print("Table of Expected Frequencies:\n", expected)
```

### Interprétation: 
Le p-value supérieur à 0.05 indique qu'il n'y a pas de relation statistiquement significative entre l'utilisation d'un dialecte dans une situation formelle pour des activités administratives et les variables examinées.

Les résultats suggèrent que l'utilisation des dialectes, particulièrement dans des contextes informels comme à la maison, est une pratique bien ancrée, alimentée par la proximité et la familiarité des interactions. Cependant, dès qu’il s'agit de situations formelles, comme les activités administratives, ils ne sont pas influencé de manière significatif par les langues parlées. Nouspouvons supposer que les participants privilégient les langues officielles, illustrant ainsi l’importance des contextes sociétaux dans la perception des langues et des dialectes.

Plus encore, des éléments comme la nationalité et le pays de résidence ne sont pas anodins : ils façonnent de manière significative les choix linguistiques, renforçant l’idée que les pratiques linguistiques sont  liées à des identités culturelles et géographiques. Le code-switching, bien que fréquemment observé, semble être plus un phénomène d’adaptabilité sociale, moins influencé par des facteurs contextuels, et plus par des préférences personnelles et des interactions sociales spécifiques.

En somme, cette recherche révèle l’influence profonde de la société et du contexte sur les choix linguistiques, une dynamique qui ne cesse d’évoluer dans un monde de plus en plus interconnecté. Elle ouvre la voie à de nouvelles interrogations sur la manière dont les langues et les dialectes s’adaptent aux défis contemporains, et comment elles façonnent, à leur tour, nos relations sociales et culturelles.
